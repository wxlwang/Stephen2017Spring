```python
#Now we're going to create the same model but with matrix calculations
#Note that X and Y are DATAFRAMES
X = sm.add_constant(X)
y = faithful.eruptions
```

```python
#First, we'll compute X'X, which is a symmetric matrix
tranX = np.dot(X.T, X) #NOT a dot product, this is matrix multiplication
print tranX
```

```python
#Now, we'll take the inverse of our new symmetric matrix
invTranX = np.linalg.inv(tranX) #linalg is the python linear algebra library
print invTranX
```

```python
#Finally, our Beta estimate for our model will be (X'X)-1 * (X'Y)
tranXY = np.dot(X.T, y)
print tranXY
```

```python
beta = np.dot(invTranX, tranXY)
print "Directly Estimating Beta:", beta
print "Estimating Beta using functions from statsmodels:", resultsW0.params.values
```

```python
ringData=np.array([[float(j) for j in e.strip().split()] for e in open(
            "C:\Users\sgbru\Desktop\IntroToDataScience\Stephen2017Spring\lab4ORingData.txt")]) #the strip and split are telling python
                                                                             #are telling python how to read the data set,
                                                                             #similar to that thing in SAS
```

```python
# fit logistic regression model
import statsmodels.api as sm
from statsmodels.formula.api import logit, glm, ols

# statsmodels works nicely with pandas dataframes
dat = pd.DataFrame(ringData, columns = ['Temperature', 'Failure'])
logit_model = logit('Failure ~ Temperature',dat).fit()
print logit_model.summary() #small p-values are good because it shows that there is, by pure chance, a relationship
                            #between our variables
#So, if p < 0.05, then we are "fairly certain" that a relationship exists between our variables \
```

```python
#So we want to counteract the process of "multiple comparisons," which we'll do using the Bonferroni comparison
# calculate predicted failure probabilities for new termperatures
x = np.linspace(50, 85, 1000)
p = logit_model.params
eta = p['Intercept'] + x*p['Temperature']
y = np.exp(eta)/(1 + np.exp(eta))
```
